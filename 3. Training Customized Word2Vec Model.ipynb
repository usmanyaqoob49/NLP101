```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gensim
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
#reading the dataset
df= pd.read_csv('pre_processed.csv')
df.head()
```

```python
df.drop('Unnamed: 0', axis= 1, inplace= True) 
```

```python
df.head()
```

As we are training our own word2vec so we have to combine these rows to form a whole corpus that can be further processed.

We have to pass each sentence to simple_preprocess() of the gensim to process the each sentence for training the model.

```python
#we will pass each tokenized sentence to simple_preprocess
complete_data = df['nltk_sentence_tokenize'].apply(lambda x: simple_preprocess(x)).tolist()
```

```python
complete_data
```

```python
len(complete_data)
```

**Now we have the data preprocessed, now we just need to make a model using same library and we will train that model using this prepared data**

```python
model= gensim.models.Word2Vec(
    #we have to select the window size---> that mean how many context words on each side of the target
    window= 10, #here 10 words will be on right of target and 10 on left

    #we will only take only those sentences having at least two words
    min_count= 2,

    #the resulting representational vector for each word
    vector_size= 100)
```

Extracting Unique Words from the whole corpus/data and making the vocabulary:

```python
model.build_vocab(complete_data)
```

Now we have to train the model using data and above settings:

```python
model.train(
    #first of all giving data
    complete_data,

    #number of sentences in the whole data/corpus
    total_examples= model.corpus_count,

    #setting epochs for the model
    epochs= 5
)
```

**Now we got our word2vec model that have convert each word to vecotr representation, lets test for some cases:**

As this dataset was for reviews we can check how many words are related to this word reviews:

```python
model.wv.most_similar('reviews')
```

**Representaional Vector of the Single Word:**

```python
model.wv['review']
```

**To check Similarilty Between two Words:**

```python
model.wv.similarity('comment', 'review')
```

```python
model.wv.similarity('review', 'python')
```

As We can see above Comment and Review were related so we had high similarity score on the other hand Review had no relation with Python which is a programming Language.

**We can get words from the correspoding Vector Representation using index_to_key:**

```python
y = model.wv.index_to_key
y
```

#### Visualization of the Vectors:

We can visualize vector to see the relation among them in more compact form but for that we have reduce the dimensions from 100 to 3d so for that we will use PCA which is a Machine Learning Technique:

```python
from sklearn.decomposition import PCA
```

As we are reducing to 3 dimension:

```python
pca = PCA(n_components=3)
```

```python
X = pca.fit_transform(model.wv.get_normed_vectors())
```

```python
X.shape
```

```python
import plotly.express as px
fig = px.scatter_3d(X[200:300],x=0,y=1,z=2, color=y[200:300])
fig.show()
```
